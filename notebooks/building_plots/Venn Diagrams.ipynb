```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
# Define the set sizes
venn = venn2(subsets=(30, 60, 10), set_labels=('A', 'B'))

# Customize appearance (optional)
venn.get_label_by_id('10').set_text('P(A)')
venn.get_label_by_id('01').set_text('P(B)')
venn.get_label_by_id('11').set_text('P(A ∩ B)')

# Display the Venn diagram
plt.title("Two-Set Venn Diagram")
plt.show()
```
